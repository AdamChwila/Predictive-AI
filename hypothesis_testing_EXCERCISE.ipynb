### Our H₀ is that true population average height is equal to value POPMEAN
### For the below code try to change: data in np.array i.e. add new records, add outliers etc. or change the popmean (tested value) or change level of alpha

```python
# Loading of packages

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

### Try to change values below and see what happens with the outcome

```python
data = np.array([175, 160, 193, 168, 180, 155, 183, 172, 179, 166])
```

### Try to change tested value and see what happens with the outcome

```python
POPMEAN = 180
```

### Try to change the significance level (alpha) and see what happens with the outcome

```python
alpha = 0.05
```

### After applying changes to the above, run the below cell and see the outcome

```python
# Basic statistics
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

# 95% confidence interval
confidence = 0.95
t_value = stats.t.ppf((1 + confidence) / 2, df=n-1)
margin_of_error = t_value * std / np.sqrt(n)
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

# Visualization
fig, ax = plt.subplots(figsize=(8, 4))

# Plot individual data points
ax.scatter(data, np.ones_like(data), color='blue', label='Data points', zorder=2)

# Plot mean
ax.plot(mean, 1, 'o', color='red', markersize=12, label='Mean', zorder=3)

# Plot 95% CI as shaded area
ax.fill_betweenx([0.9, 1.1], ci_lower, ci_upper, color='skyblue', alpha=0.3, label='95% CI', zorder=1)

# Formatting
ax.set_yticks([])
ax.set_xlabel('Value')
ax.set_title('Mean and 95% Confidence Interval')
ax.legend()
ax.set_xlim(min(data)-2, max(data)+2)

plt.show()

print(f"Sample mean: {mean:.2f}")
print(f"95% CI: ({ci_lower:.2f}, {ci_upper:.2f})")

# One-sample t-test (e.g., test if mean differs from 90)

##### Try to change popmean or alpha values below and see what happens with the outcome ###
t_stat, p_value = stats.ttest_1samp(data, popmean=POPMEAN)

if p_value < alpha:
    decision = f"p-value = {p_value:.4f} < {alpha}, so we reject H₀."
else:
    decision = f"p-value = {p_value:.4f} ≥ {alpha}, so we do not reject H₀."

print("p-value tells us how well our data fits the tested hypothesis H₀.")
print(f"{decision}")
```
